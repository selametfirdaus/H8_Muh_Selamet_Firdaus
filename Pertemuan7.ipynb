```python
# we can pass mutliple criteria in the same line. 
# let's filter for AreaNAme = Asia and RegName = Southern Asia

#df_can[((df_can['Continent']=='Asia') | (df_can['Continent']=='Europe')) & (df_can['Region']=='Southern Asia')]
#df_can[((df_can['Continent']=='Asia') | (df_can['Continent']=='Europe'))]
df_can[((df_can['Continent']=='Asia') | (df_can['Continent']=='Europe')) & (df_can['DevName']=='Developed regions')]
#df_can[(df_can['Continent'=['Asia','Europe']) & (df_can['DevName']=='Developed regions')]

# note: When using 'and' and 'or' operators, pandas requires we use '&' and '|' instead of 'and' and 'or'
# don't forget to enclose the two conditions in parentheses
```
