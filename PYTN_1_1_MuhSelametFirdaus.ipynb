# LATIHAN 1 Kenaikan jumlah kasus dari 2011 ke 2016

```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

```python
df_can = pd.read_csv('london_crime_by_lsoa.csv')

print('Data downloaded and read into a dataframe!')
```

```python
df_can.describe()
```

```python
# we are using the inline backend
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
jumlah = df_can[
    (df_can["year"] >= 2011) &
    (df_can["year"] <= 2016)
].groupby(["year"])["value"].sum()
```

```python
print(jumlah)
```

```python
jumlah.plot()
```

```python
jumlah.index = jumlah.index.map(int)
jumlah.plot(kind='line')

plt.title('Kenaikan Jumlah Kasus')
plt.ylabel('Jumlah Kasus')
plt.xlabel('Tahun')

plt.show() 
```

# LATIHAN 2 Tindak Kriminal Perdaerah pada tahun 2016

```python
df_can.head(10)
```

```python
df_can[
    (df_can["year"] == 2016)
].groupby(["borough","major_category"])["value"].sum()
```

```python
jumlah2 = df_can[
    (df_can["year"] == 2016)
].groupby(["borough"])["value"].sum()
```

```python
print(jumlah2)
```

```python
jumlah2.plot()
```

```python
jumlah2.plot(kind='line', figsize=(30, 15))

plt.title('Jumlah Kasus Pada Tahun 2016')
plt.ylabel('Jumlah Kasus')
plt.xlabel('Nama Daerah')

plt.show() 
```

```python
jumlah2.plot(kind='bar', figsize=(20, 6))

plt.xlabel('Nama Daerah') 
plt.ylabel('Jumlah') 
plt.title('Jumlah Kasus Pada Tahun 2016') 

plt.show()
```

```python
jumlah2.plot(kind='barh', figsize=(12, 12), color='steelblue')
plt.xlabel('Jumlah Kasus')
plt.title('Jumlah Kasus Pada Tahun 2016')

plt.show()
```
